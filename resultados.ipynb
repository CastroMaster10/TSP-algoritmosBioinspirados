### Visualizacion de todos los 150 direcciones

```python
def visualizarLugares(latitudes,longitudes):

    coordenadas = [(latitudes[i],longitudes[i]) for i in range(len(latitudes))]

    mapa =  folium.Map(location=coordenadas[0],zoom_start=13)#Creamos el mapa empezando por el primer lugar

    for coord in coordenadas:
        #Usamos la libreria literal_eval para convertir las cadenas de texto en una tupla
        folium.Marker(coord).add_to(mapa)
    
    return mapa
```

```python
latitudes,longitudes = (direcciones["Latitud"].tolist(),direcciones["Longitud"].tolist())
mapa = visualizarLugares(latitudes,longitudes)
```
